**LGM - Task2 - Image To Pencil Sketch Using Python.**
 

Import required libraries 

```python
from google.colab.patches import cv2_imshow
from google.colab import files 
import cv2
```

Upload the image from system

```python
uploaded = files.upload()
```

Display image using imread function. This function reads the image.

```python
original_image = cv2.imread('Tweety.jpg')
cv2_imshow(original_image)
```

Using cvtColor function of cv2 to convert image to grayscale which provides black and white pixels.

```python
gray_img = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

Inverting the image using bitwise_not to make brighter regions lighter and vice versa.

```python
inverted_img = cv2.bitwise_not(gray_img)
cv2_imshow(inverted_img)
```

Using GaussianBlur function from cv2 make the image smooth which creates thin lines.

```python
img_smooth = cv2.GaussianBlur(inverted_img,(21,21),sigmaX=0, sigmaY=0)
cv2_imshow(img_smooth)
```

Divide the grayscale value of the image by inverse blurred image value.  

```python
def dodgeV2(x,y):
  return cv2.divide(x, 255 - y, scale=256)
```

Create the pencil sketch 

```python
final_img = dodgeV2(gray_img, img_smooth)
cv2_imshow(final_img)
```
